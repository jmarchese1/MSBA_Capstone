**Preprocessing financial data**

```python
model_data.head(2) #normalize the data
```

```python
#need to drop or impute rows with missing values...

model_data.isnull().sum()
```

```python
clean_model_data = model_data.dropna()
len(clean_model_data) #still enough data to make predictions...
```

```python
final_model_data = clean_model_data[clean_model_data["tuition"] > 0]
len(final_model_data)
```

```python
#splitting data into training and test sets
from sklearn.model_selection import train_test_split
X = final_model_data.drop(["tuition"], axis = 1)
y = final_model_data["tuition"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(X_train, y_train)
```

```python
preds = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, preds) #model predicts with in 16% of the actual tuition of the school
```

```python
plt.title("predictions vs true values")
plt.scatter(y_test, preds, alpha=0.7, color = "green")
plt.xlabel("true values")
plt.ylabel("predictions")
```

```python
#going to build a model without any outlier
from scipy.stats import zscore
```

```python
z_scores = np.abs(zscore(X_train))

X_train_no_outliers = X_train[(z_scores < 3).all(axis = 1)]
y_train_no_outliers = y_train[(z_scores < 3).all(axis = 1)]
```

```python
model_1 = LinearRegression()
model_1.fit(X_train_no_outliers, y_train_no_outliers)
```

```python
model_1_preds = model_1.predict(X_test)
```

```python
mean_absolute_percentage_error(y_test, model_1_preds)
```

**with in 7% of true value** ü•á

```python
plt.scatter(x = y_test, y = model_1_preds, color = "green", label = "new", alpha = 0.3)
plt.scatter(x = y_test, y = preds, color = "r", label = "old", alpha = 0.3)
plt.legend()
plt.xlabel("true values")
plt.ylabel("predictions")
plt.title("old model vs new model")

```
